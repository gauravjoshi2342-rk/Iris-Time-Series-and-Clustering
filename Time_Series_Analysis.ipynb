```python
#importing the python libaries for analysis and visulization
import pandas as pd #best for handle the time-based data
import numpy as np #to work on mathmetical calculations
import matplotlib.pyplot as plt #plotting the graphs to see trends visually
from statsmodels.tsa.seasonal import seasonal_decompose #it breaks raw data into 3 diffrent parts (Trends,Seasonality,Residuals)

#datast generation - creating a synthetic time series to demonstrate trends and seasonality
dates = pd.date_range(start='2026-01-01', periods=100, freq='D')
values = np.linspace(0, 10, 100) + np.sin(np.linspace(0, 20, 100)) + np.random.normal(0, 0.2, 100)
df_ts = pd.DataFrame({'Date': dates, 'Value': values}).set_index('Date')

#visualization - plotting raw data to identify the intial data distribution and noise.
plt.figure(figsize=(10, 5))
plt.plot(df_ts)
plt.title('Time Series Data')
plt.savefig('Original_Data.png') #saving the image of original data
plt.show()

# 3. Decomposition - breaking down the series into trend,seasonality and residuals.
result = seasonal_decompose(df_ts['Value'], model='additive', period=30)
result.plot()
plt.savefig('Decomposition.png') #saving the image of decomposition
plt.show()

# 4. Moving Average - smoothing the series to reduce noise and highlight long term trend.
df_ts['MA'] = df_ts['Value'].rolling(window=7).mean()
plt.figure(figsize=(10, 5))
plt.plot(df_ts['Value'], label='Original')
plt.plot(df_ts['MA'], label='Moving Average', color='red')
plt.legend()
plt.savefig('Moving_Average.png')
plt.show() #showing the figure of moving average by py library(matplotlib)
```
